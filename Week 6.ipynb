```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
#线性空间函数
x = np.linspace(-3.0, 3.0, 100)

# Immediately, the result is given to us.  An array of 100 numbers equally spaced from -3.0 to 3.0.
print(x)

# We know from numpy arrays that they have a `shape`, in this case a 1-dimensional array of 100 values
print(x.shape)

# and a `dtype`, in this case float64, or 64 bit floating point values.（数据类型对象 ( ) ）
print(x.dtype)
```

```python
#考虑tf.Tensor的方式与你考虑numpy.array的方式相同。它是由它的形状描述的，在这种情况下，只有1维的100个值。而且它有一个dtype，在这种情况下，是float32。但是与numpy.array不同的是，这里没有打印出任何数值!这是因为它实际上还没有计算出它的值。相反，它只是指一个已经被添加到Tensorflow默认计算图中的tf.Operation的输出。该操作的结果就是我们所返回的张量。
x = tf.linspace(-3.0, 3.0, 100)
print(x)
```

```python
'''
让我们试着检查一下表格。我们可以请求 "默认 "图，在那里我们所有的操作都被添加了。
'''
g = tf.get_default_graph()
```

```python
'''
从这个图中我们可以得到一个所有被添加的操作的列表，并打印出它们的名字
'''
[op.name for op in g.get_operations()]
'''
因此，Tensorflow对我们的每个操作都进行了命名，以大体上反映它们在做什么。
有几个参数都以LinSpace为前缀，然后最后一个是操作，它接收所有的参数并为linspace创建一个输出。'''
```

```python
'''
我们可以通过向图询问张量的名称来请求任何操作的输出，这就是张量。'''
g.get_tensor_by_name('linspace/Slice' + ':0')
'''
我所做的是要求获得来自操作 "LinSpace "的tf.Tensor。所以记住，tf.Operation的结果是一个tf.Tensor。
请记住，这与我们之前创建的张量x是同一个名字。
'''
```

```python
'''
为了在tensorflow中实际计算任何东西，我们需要创建一个tf.Session。
这个会话负责评估tf.Graph。让我们来看看这是如何工作的。'''
# We're first going to create a session:创建一个session
sess = tf.Session()

# Now we tell our session to compute anything we've created in the tensorflow graph.
computed_x = sess.run(x)
print(computed_x)

# Alternatively, we could tell the previous Tensor to evaluate itself using this session:
computed_x = x.eval(session=sess)
print(computed_x)

# We can close the session after we're done like so:
sess.close()
```

```python
'''
我们也可以明确地告诉会话我们要管理哪个图。这里是指前面那个默认的图'''
sess = tf.Session(graph=g)
sess.close()
#默认情况下，它抓取的是默认图。但我们可以像这样创建一个新的图形。
g2 = tf.Graph()
```

```python
'''然后只在我们的会话中使用这个图。
为了简化事情，由于我们将在iPython的交互式控制台中工作，我们可以创建一个tf.InteractiveSession。'''
sess = tf.InteractiveSession()
x.eval()
'''
现在我们不需要明确地告诉eval函数我们的会话。在接下来的讲座中，我们会让这个会话开放。'''
```

```python
'''
让我们现在尝试一组操作。我们将尝试创建一个高斯曲线。
这应该类似于一个归一化的直方图，其中大部分数据的中心都在平均值0的周围。'''
```

```python
#help(plt.imshow)
#help(np.exp)
# help(plt.imread)
# help(os.path.join)
```

```python
'''The 1 dimensional gaussian takes two parameters, the mean value, and the standard deviation, which is commonly denoted by the name sigma.
一维高斯有两个参数，平均值和标准差，通常用sigma这个名字表示。
'''
mean = 0.0
sigma = 1.0

# Don't worry about trying to learn or remember this formula.  I always have to refer to textbooks or check online for the exact formula.
z = (tf.exp(tf.negative(tf.pow(x - mean, 2.0) /
                   (2.0 * tf.pow(sigma, 2.0)))) *
     (1.0 / (sigma * tf.sqrt(2.0 * 3.1415))))
'''
就像以前一样，令人惊讶的是，我们实际上并没有计算任何东西。我们只是给Tensorflow的图添加了一堆操作。每当我们想要这个操作的值或输出时，我们就必须明确询问我们感兴趣的图的部分，然后才能看到其结果。
由于我们已经创建了一个交互式会话，我们应该只需要说出我们感兴趣的张量的名称，然后调用计算函数。'''
```

```python
res = z.eval()
plt.plot(res)
# if nothing is drawn, and you are using ipython notebook, uncomment the next two lines:
#%matplotlib inline
#plt.plot(res)
```

```python
'''
让我们试着创建一个二维高斯--基本上是我们刚才所做的事情的二维版本。
这可以通过将一个向量与它的转置相乘来实现。如果你不熟悉矩阵数学，我将回顾一下几个重要的概念。
这大约是神经网络工作的98%，所以如果你不熟悉这个，那么请你跟我一起完成这个，就会很顺利了。
首先，要将两个矩阵相乘，它们的内部维度必须一致，而产生的矩阵将具有外部维度的形状。
因此，假设我们有两个矩阵，X和Y。为了使我们能够将它们相乘，X的列必须与Y的行一致。我试着这样记住它。
但我们的矩阵实际上是一个矢量，或一个一维矩阵。这意味着它的维度是N×1，所以要把它们相乘，我们就得'''
# Let's store the number of values in our Gaussian curve.
ksize = z.get_shape().as_list()[0]

# Let's multiply the two to get a 2d gaussian
z_2d = tf.matmul(tf.reshape(z, [ksize, 1]), tf.reshape(z, [1, ksize]))

# Execute the graph
plt.imshow(z_2d.eval())
```
