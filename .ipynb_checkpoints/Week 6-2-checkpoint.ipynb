```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.markers import MarkerStyle
from matplotlib.transforms import Affine2D
from matplotlib.text import TextPath
from matplotlib.colors import Normalize

SUCCESS_SYMBOLS = [
    TextPath((0, 0), "@"),
    TextPath((0, 0), "😒"),
    TextPath((0, 0), "☺"),
]

N = 25
np.random.seed(42)
skills = np.random.uniform(5, 80, size=N) * 0.1 + 5
# rotation
takeoff_angles = np.random.normal(0, 90, N)
# color
thrusts = np.random.uniform(size=N)
successful = np.random.randint(0, 3, size=N)
positions = np.random.normal(size=(N, 2)) * 5
data = zip(skills, takeoff_angles, thrusts, successful, positions)

cmap = plt.colormaps["plasma"]
fig, ax = plt.subplots()
fig.suptitle("Throwing success", size=14)
for skill, takeoff, thrust, mood, pos in data:
    t = Affine2D().scale(skill).rotate_deg(takeoff)
    m = MarkerStyle(SUCCESS_SYMBOLS[mood], transform=t)
    ax.plot(pos[0], pos[1], marker=m, color=cmap(thrust))
fig.colorbar(plt.cm.ScalarMappable(norm=Normalize(0, 1), cmap=cmap),
             ax=ax, label="Normalized Thrust [a.u.]")
ax.set_xlabel("X position [m]")
ax.set_ylabel("Y position [m]")

plt.show()
```

```python
# The 1 dimensional gaussian takes two parameters, the mean value, and the standard deviation, which is commonly denoted by the name sigma.
mean = 0.0
sigma = 1.0

# Don't worry about trying to learn or remember this formula.  I always have to refer to textbooks or check online for the exact formula.
z = (tf.exp(tf.negative(tf.pow(x - mean, 4.0) /
                   (2.0 * tf.pow(sigma, 4)))) *
     (1.0 / (sigma * tf.sqrt(2.0 * 3.1415))))
# 这段代码计算了一个正态分布的概率密度函数。其中x是输入的自变量（可以是一个数，也可以是一个向量），mean是正态分布的均值，sigma是正态分布的标准差。z表示对应于输入x的概率密度函数的值。具体地，z的计算方式是将x代入正态分布的公式，计算出对应于x的概率密度函数的值。
```

```python
res = z.eval()
plt.plot(res)
```

```python
# Let's store the number of values in our Gaussian curve.
data1 = np.random.random([6, 50])
ksize = z.get_shape().as_list()[0]

# Let's multiply the two to get a 2d gaussian
z_2d = tf.matmul(tf.reshape(z, [ksize, 1]), tf.reshape(z, [1, ksize]))

# Execute the graph
plt.imshow(z_2d.eval())
'''
这段代码首先通过 z.get_shape().as_list()[0] 获取了张量 z 的第一个维度大小，即 ksize。
然后使用 tf.reshape 将 z 转换成形状为 (ksize, 1) 和 (1, ksize) 的两个矩阵，并使用 tf.matmul 计算它们的乘积，得到一个二维的矩阵 z_2d。

最后，使用 plt.imshow(z_2d.eval()) 将 z_2d 可视化为一张灰度图像。由于 z 是一个高斯分布， z_2d 实际上是一个二维高斯分布的密度图。
'''
```
